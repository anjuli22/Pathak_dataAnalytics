```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

<h3> input from user file name and date

```python
#All Imports collected 
import os

file_path=os.getcwd()

file_data = pd.read_csv(file_path + '/file_list.csv')
print(file_data)
```

```python
def file_name_ip():
    file_name= str((input("Enter file name for which we want to check the regression").strip()))
    c=0
    while c==0:
        if file_name in file_data.values:
            c=1
            return file_name
        else:
            file_name= str((input("File name entered wrong. Enter file name for which we want to check the regression").strip()))
        
    return file_name
```

```python
#input date and format check

import re

def date_inp():
    date_file_name = str((input("Enter date - Format (dd-mm-yy)").strip()))
    d=0
    while d==0:
        if re.match(r"^\d{2}[-]\d{2}[-]\d{2}$",date_file_name):
            d=d+1
            return date_file_name
        else:
            date_file_name = str((input("Entered wrong date, please try again:(dd-mm-yy)").strip()))

    return date_file_name
```

```python
t=0
while t==0:
    try:
        name=file_name_ip()
        dt=date_inp()
        file= dt+'\\'+name+'.csv'
        data_v1=pd.read_csv(file)
        #data_v1.head()
        t+=1
        print('file read successfully')
    except:
        t==1
        print('the date range you have mentioned is not present, try again')
        #print('g')
    
```

Train and test data 

```python
X = data_v2.filter(['pvr','volume','fl916','m30exp','net_count'],axis=1)
y = data_v2['ltp']
X.head()
```

```python
y.head()
```

```python
X.dtypes

X['fl916'].astype('float')
```

```python
#split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)

print(len(X_train.index))
print(len(X_test.index))
print(len(y_train.index))
print(len(y_test.index))

# train and test data should be 80:20 ratio
```

<h1> Linear Regression models 

<H3> Model 1 - no change in fl916

```python
from sklearn.linear_model import LinearRegression


lm = LinearRegression()  # create object for the class
model_1=lm.fit(X_train, y_train)  # perform linear regression
Y_pred = lm.predict(X_test)  # make predictions

print(Y_pred)
#Y_pred predict dependent values
```

```python
#calculating accuracy of the model :


from sklearn.metrics import r2_score
r2_1=r2_score(y_test,Y_pred)
print('R-square of linear model ', r2_1)
```

Our Linear Regression Model is 96.04 % accurate

```python
#plotting regression model

plt.figure(figsize=(15,10))
plt.scatter(y_test, Y_pred, color='lime')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Linear Model for raw data')
plt.show()
```

```python
dif_df= pd.DataFrame({'updatetime':data_v2['updatetime'],'ltp_dif':0,'ltp':data_v2['ltp']})
dif_df
```

```python
l=len(dif_df)

for x in range(l):
    if x==0:
        prev=dif_df.iloc[0,2]
        #print(prev)
    else:
        dif_df.iloc[x,1]=dif_df.iloc[x,2]-prev


    
```

```python
dif_df
```

```python

dif_df['Differnce_flag'] = dif_df['ltp_dif'].apply(lambda x: 'N' if x <0 else 'P')
dif_df['abs_ltp_dif']=abs(dif_df['ltp_dif'])

#sort desc desc_dif_df

desc_dif_df= dif_df.sort_values(by='abs_ltp_dif', ascending=False)
```

```python
mean_dif=desc_dif_df['abs_ltp_dif'].mean()
print(mean_dif)
```
